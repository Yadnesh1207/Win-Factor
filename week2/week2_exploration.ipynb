```python
data = pd.read_csv('clean_data.csv')
```

```python
for col in data.columns:
    uni = data[col].nunique()
    print("unique values in",col,"=",uni)
```

```python
male_data = data[data['gender'] == 'male']
female_data = data[data['gender'] == 'female']
```

```python
male_data.drop(['gender'], axis=1, inplace=True)
female_data.drop(['gender'], axis=1, inplace=True)
```

```python
male_data.to_csv('male_data.csv',index=False)
female_data.to_csv('female_data.csv',index=False)
```

```python
male_data.head()
```

```python
for col in male_data.columns:
    uni = male_data[col].nunique()
    print("unique values in",col,"=",uni)
```

```python
sns.pairplot(male_data, vars= [
    'runs_scored', 'balls_faced',
    'wickets_taken',
    'dot_balls_as_batsman', 'dot_balls_as_bowler',
    'catches_taken','fantasy_score_bowling', 'fantasy_score_total'
],  kind = 'scatter' , diag_kind = 'kde')
plt.show()
```

```python
numeric_cols = [
    'runs_scored', 'balls_faced', 'fours_scored', 'sixes_scored',
    'wickets_taken', 'runs_conceded', 'maidens',
    'dot_balls_as_batsman', 'dot_balls_as_bowler',
    'catches_taken','fantasy_score_batting','fantasy_score_bowling', 'fantasy_score_total'
]



# Calculate correlation matrix
corr = male_data[numeric_cols].corr()


# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap with the mask
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title("Pairwise Correlation Heatmap")
plt.tight_layout()
plt.show()
```
